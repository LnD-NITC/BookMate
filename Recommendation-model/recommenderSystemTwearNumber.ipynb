```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings'] >= 200].sort_values('avg_rating', ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books, on='Book-Title')
popular_df
```

```python
popular_df = popular_df.drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_rating']]
```

```python
popular_df.count()
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 20
top_users = x[x].index
```

```python
top_users.size
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(top_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] > 15
top_books = y[y].index
```

```python
top_books.size
```

```python
top_books
```

```python
top_books_df = books[books['Book-Title'].isin(top_books)][['Book-Title', 'ISBN']]
```

```python
top_books_df
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(top_books)]
```

```python
final_ratings
```

```python
test = final_ratings.drop_duplicates(subset='ISBN', keep='first')
test.count()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
np.save('similarity_scores.npy', similarity_scores)  
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        # item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend("Misery")
```
